# **EDA for Marvel & DC Character**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
marvel = pd.read_csv("charcters_stats.csv")
```

```python
marvel.info()
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
mar_good = marvel[marvel['Alignment']=='good']
```

```python
mar_good.head()
```

```python
mar_good.sort_values(by=['Speed'], ascending=False).head()
```

```python
good_max_power = mar_good[mar_good['Power'] == 100].head()
good_max_power
```

```python
good_max_power = mar_good.sort_values(by=['Total'], ascending=False).head()
good_max_power
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=list(good_max_power['Name'])[0:6], y=list(good_max_power['Total'])[0:6], color='g')
plt.show()
```

```python
good_speed = mar_good.sort_values(by=['Speed'], ascending=False).head()
good_speed
```

```python
plt.figure(figsize=(10,5))
sns.histplot(list(mar_good['Speed']))
plt.xlabel('speed')
plt.show()
```

```python
mar_bad = marvel[marvel['Alignment'] == 'bad']
mar_bad
```

```python
mar_bad.sort_values(by=['Speed'], ascending=False).head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(list(mar_bad['Combat']))
plt.xlabel('Combat')
plt.show()
```
